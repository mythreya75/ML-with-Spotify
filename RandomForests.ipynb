```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("out.csv")
```

```python
dataset["release_year"][0] >= 2016 and dataset["release_year"][0] < 2020
```

```python
dataset["rap_era"] = ["None"]*len(dataset)
for i in range(len(dataset)):
    if dataset["release_year"][i] >= 1986 and dataset["release_year"][i] < 1992:
        dataset["rap_era"].replace(["None"], ["1986-1991"])
    elif dataset["release_year"][i] >= 1992 and dataset["release_year"][i] < 1996:
        dataset["rap_era"].replace(["None"], ["1992-1995"])
    elif dataset["release_year"][i] >= 1996 and dataset["release_year"][i] < 2000:
        dataset["rap_era"].replace("None", "1996-1999")
    elif dataset["release_year"][i] >= 2000 and dataset["release_year"][i] < 2004:
        dataset["rap_era"].replace("None", "2000-2003")
    elif dataset["release_year"][i] >= 2004 and dataset["release_year"][i] < 2008:
        dataset["rap_era"].replace("None", "2004-2007")
    elif dataset["release_year"][i] >= 2008 and dataset["release_year"][i] < 2012:
        dataset["rap_era"].replace("None", "2008-2011")
    elif dataset["release_year"][i] >= 2012 and dataset["release_year"][i] < 2016:
        dataset["rap_era"].replace("None", "2012-2015")
    elif dataset["release_year"][i] >= 2016 and dataset["release_year"][i] < 2020:
        dataset["rap_era"].replace("None", "2016-2019")
    elif dataset["release_year"][i] >= 2020:
        dataset["rap_era"].replace(["None"], ["2020-Present"])    
    print(dataset["rap_era"][i])
```

```python
dataset
```

```python
X = dataset.iloc[:,1:13].values
y = dataset.iloc[:, 13].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000, random_state=530)
classifier.fit(X_train, y_train)
prediction = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, prediction))
print(classification_report(y_test, prediction))
print(accuracy_score(y_test, prediction))
```
